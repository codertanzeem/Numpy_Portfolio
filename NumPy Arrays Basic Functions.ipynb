```python
#built-in methods(popular)
#***np.arange***

np.arange(0,10)
```

```python
np.arange(0,50,2) #start at 0, stop at 50, with step size of 2
```

```python
#***np.zeros***
np.zeros(5) #1D vector
```

```python
np.zeros((5,5)) #2D vector
```

```python
#***np.ones***
np.ones(5)
```

```python
#***np.linspace

np.linspace(0,10,3) #start at 0, stop at 10, with 3 evenly spaced points
```

```python
np.linspace(0,10,11)
```

```python
len(np.linspace(0,10,11))
```

```python
#***np.eye*** creates an identity matrix

np.eye(5)
```

```python
#***np.random.rand*** ----> gives the number between 0 and 1 where zero is not included while 1 is
#all the numbers are in uniform distribution which means all have equal likelyhood to be an output
np.random.rand(1)
```

```python
np.random.rand(5,2)
```

```python
#***np.random.randn***

np.random.randn(5)
```

```python
#***np.tandom.randint*** ----> lower number is inclusive here and higher is exclusive

np.random.randint(0,101,5)
```

```python
#***np.random.seed***
#In NumPy, the numpy.random.seed() function sets the seed for generating random numbers.
#This allows the user to specify a value that will ensure the same sequence of random
#numbers is generated each time the code is run. This is known as deterministic behavior
#and allows for reproducibility.

np.random.seed(42)
np.random.rand(4)
```
